```python
# NLP
# Data Preprocessing and visualizations

# """ 
# The processes and steps involved in preprocessing of Text data for NLP are:
# 1. Sentence Segmentation
# 2. Word Tokenization
# 3. Stemming
# 4. Lemmatization
# 5. Stop Words
# """
```

```python
# Sentence segmentation
       # Separate sentences
    
import nltk

token_sent = nltk.sent_tokenize(file)
print(token_sent)
```

```python
token_word = nltk.word_tokenize(file)
print(token_word)
```

```python
token_word;
```

```python
# Remove punctuations and numerics and make tokens lowercase

words = [word.lower() for word in token_word if word.isalpha()]

print(words)

```

```python
# Stemming and Lemmatizing
# Shortening the words by cutting off the tails for eg. eating ==> eat, jumps ==> jump
# Lemmatizing ===> more concious

# PorterStemmer is one of the most used
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in words]
stemmed_text = ' '.join(stemmed_words)
print(stemmed_text)
```

```python
len(stemmed_text)
```

```python
# Remove Stop Words
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
y = []
for i in stemmed_text:
    if i not in stopwords.words('english'):
        y.append(i)

    
```

```python
from nltk import FreqDist
from nltk.tokenize import word_tokenize

# # Tokenize the filtered text into individual words
tokens = word_tokenize(stemmed_text)

# Calculate the frequency distribution of the words
freq_dist = FreqDist(tokens)

# Print the most common words and their frequencies
top_words = 12  # top 12 frequently used words
for word, freq in freq_dist.most_common(top_words):
    print(word, '-', freq)
```
